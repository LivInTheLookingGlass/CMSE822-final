```python
import matplotlib.pyplot as plt
import numpy as np
import csv

# close all the figures, if open from previous commands
plt.close('all') 

########### Plot Data ###################
 
N = []
n = []
t = []
average_time = []

# open file for reading
with open('results_default_strong.csv') as csvDataFile:

    # read file as csv file 
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        N.append(row[1])
        n.append(row[2])
        t.append(row[3])
        average_time.append(row[6])

N = np.array(N)
n = np.array(n)
t = np.array(t)
average_time = np.array(average_time)
        
N = np.delete(N, 0)
n = np.delete(n, 0)
t = np.delete(t, 0)
average_time = np.delete(average_time, 0)


N = [int(i) for i in N]
n = [int(i) for i in n]
t = [int(i) for i in t]
average_time = [float(i) for i in average_time]


########### Split Into Ranges ###################
#1 Pink 
default_strong_32_N = N[0:8]
default_strong_32_n = n[0:8]
default_strong_32_t = t[0:8]
default_strong_32_average_time = average_time[0:8]
#2 Red
default_strong_64_N = N[8:16]
default_strong_64_n = n[8:16]
default_strong_64_t = t[8:16]
default_strong_64_average_time = average_time[8:16]
#3 Orange
default_strong_128_N = N[16:24]
default_strong_128_n = n[16:24]
default_strong_128_t = t[16:24]
default_strong_128_average_time = average_time[16:24]
#4 Yellow
default_strong_256_N = N[24:32]
default_strong_256_n = n[24:32]
default_strong_256_t = t[24:32]
default_strong_256_average_time = average_time[24:32]
#5 Green
default_strong_512_N = N[32:40]
default_strong_512_n = n[32:40]
default_strong_512_t = t[32:40]
default_strong_512_average_time = average_time[32:40]
#6 Blue
default_strong_1024_N = N[40:48]
default_strong_1024_n = n[40:48]
default_strong_1024_t = t[40:48]
default_strong_1024_average_time = average_time[40:48]
#7 Purple
default_strong_2048_N = N[48:56]
default_strong_2048_n = n[48:56]
default_strong_2048_t = t[48:56]
default_strong_2048_average_time = average_time[48:56]


########### Line style and Marker ###################
plt.plot(default_strong_32_t,default_strong_32_average_time, 'o-',color = 'pink',label='32')  
plt.plot(default_strong_64_t,default_strong_64_average_time, 'o-',color = 'red',label='64')  
plt.plot(default_strong_128_t,default_strong_128_average_time, 'o-',color = 'orange',label='128')  
plt.plot(default_strong_256_t,default_strong_256_average_time, 'o-',color = 'yellow',label='256')  
plt.plot(default_strong_512_t,default_strong_512_average_time, 'o-',color = 'green',label='512')  
plt.plot(default_strong_1024_t,default_strong_1024_average_time, 'o-',color = 'blue',label='1024')  
plt.plot(default_strong_2048_t,default_strong_2048_average_time, 'o-',color = 'purple',label='2048')  


# ############ Legend ####################
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
# #plt.ylim(0,5)
plt.yscale('log')

# #### Lable and Grid ####################
plt.xlabel('Number of Threads').set_fontsize(14) # x label 
plt.ylabel('Time (s)').set_fontsize(14) # y label
# plt.grid(True,which="both", linestyle='--')
plt.title('2D Burgers Strong Scaling (Default)').set_fontsize(16)
# plt.show()
```
