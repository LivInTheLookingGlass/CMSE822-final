```python
import matplotlib.pyplot as plt
import numpy as np
import csv

# close all the figures, if open from previous commands
plt.close('all') 

########### Plot Data ###################
 
a = []
b = []

# open file for reading
with open('results_static.csv') as csvDataFile:

    # read file as csv file 
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        a.append(row[4])
        b.append(row[7])

c = np.array(a)
d = np.array(b)
        
c = np.delete(c, 0)
d = np.delete(d, 0)


e = [int(i) for i in c]
f = [float(i) for i in d]

########### Line style and Marker ###################
plt.plot(e,f, 'o',color = 'teal',label='Static')  

############ Legend ####################
plt.legend(loc="best") #show legend
#plt.ylim(0,5)

#### Lable and Grid ####################
plt.xlabel('Chunk Size').set_fontsize(14) # x label 
plt.ylabel('Time (s)').set_fontsize(14) # y label
plt.grid(True,which="both", linestyle='--')
plt.title('Chunk Size vs. Time').set_fontsize(16)
plt.show()
```
