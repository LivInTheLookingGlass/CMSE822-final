```python
import numpy as np
import matplotlib.pyplot as plt

import csv

# close all the figures, if open from previous commands
plt.close('all') 

########### Plot Data ###################
 
# first cell data
a = []

# second cell data
b = []

# open file for reading
with open('results_static.csv') as csvDataFile:

    # read file as csv file 
    csvReader = csv.reader(csvDataFile)
    for row in csvReader:
        # add cell [0] to list of dates
        a.append(row[4])

        # add cell [1] to list of scores
        b.append(row[7])

c = np.array(a)
d = np.array(b)
        
c = np.delete(c, 0)
d = np.delete(d, 0)


e = [int(i) for i in c]
f = [float(i) for i in d]
#print(e)
#print(f)
#print(d)
########### Line style and Marker ###################
plt.plot(e,f, 'o r')  
#plt.plot(Num_Cores,Val_Pi, '--g', label='Peak Performance') 
#plt.plot(Num_Cores,True_Pi, '-g', label='Peak Performance') 

############ Legend ####################
#plt.legend(loc="best") #show legend
#plt.ylim(0,5)

#### Lable and Grid ####################
plt.xlabel('Number of Threads').set_fontsize(14) # x label 
plt.ylabel('Time (s)').set_fontsize(14) # y label
plt.grid(True,which="both", linestyle='--')
plt.title('Strong Scaling').set_fontsize(16)
plt.show()
```
